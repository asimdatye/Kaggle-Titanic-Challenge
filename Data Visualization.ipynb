```python
import os
import pandas as pd
from plotnine import *
import seaborn as sns
import matplotlib.pyplot as plt

os.getcwd()

train = pd.read_csv('data/train.csv')
```

```python
pd.DataFrame.head(train)
```

```python
(ggplot(train, aes(x='Age', y='Fare', fill='factor(Pclass)'))
 + geom_point()
 + facet_wrap('~ Survived')
 + stat_smooth(method = 'lm')
 + theme_bw())
```

```python
(ggplot(train, aes(x = 'Pclass', fill='factor(Pclass)'))
 + geom_bar()
 + facet_wrap('~ Sex + Survived')
 + theme_bw())
```

```python
(ggplot(train, aes(x = 'Pclass', fill='factor(Pclass)'))
 + geom_bar()
 + facet_wrap('~ Sex + Survived')
 + theme_bw())
```

```python
(ggplot(train, aes(x='factor(Pclass)', y='Age', fill='factor(Pclass)'))
 + geom_boxplot()
 + facet_wrap('~ Survived')
 + theme_bw())
```
